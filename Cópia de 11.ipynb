```python
precoP1 = float(input("Digite o preço do primeiro produto: "))
precoP2 = float(input("Digite o preço do segundo produto: "))
precoP3 = float(input("Digite o preço do terceiro produto: "))

if precoP1 < precoP2:
    if precoP1 < precoP3:
        print("Você deve comprar o primeiro produto!")
    else:
        print("Você deve comprar o terceiro produto!")
else:
    if precoP2 < precoP3:
        print("Você deve comprar o segundo produto!")
    else:
        print("Você deve comprar o terceiro produto!")
```
